# Exploratory Data Analysis

```python
data.columns
```

```python
# Count the number of matches played per season
matches_per_season = data['season'].value_counts().sort_index()

# Plot the bar chart
matches_per_season.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.xlabel('Season')
plt.ylabel('Number of Matches Played')
plt.title('Number of Matches Played per Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# total runs scored or wickets taken per season.
# Extract the columns representing runs scored or wickets taken, along with the 'season' column
runs_wickets_data = data[['season', 'home_runs', 'away_runs', 'home_wickets', 'away_wickets']]

# Group the data by the 'season' column and sum the runs scored or wickets taken for each season
runs_per_season = runs_wickets_data.groupby('season')[['home_runs', 'away_runs']].sum()
wickets_per_season = runs_wickets_data.groupby('season')[['home_wickets', 'away_wickets']].sum()

# Plot the bar chart for total runs scored per season
runs_per_season.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'salmon'])
plt.xlabel('Season')
plt.ylabel('Total Runs Scored')
plt.title('Total Runs Scored per Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot the bar chart for total wickets taken per season
wickets_per_season.plot(kind='bar', figsize=(10, 6), color=['lightgreen', 'orange'])
plt.xlabel('Season')
plt.ylabel('Total Wickets Taken')
plt.title('Total Wickets Taken per Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the number of wins for each team
wins_per_team = data['winner'].value_counts()

# Plot the bar chart
wins_per_team.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Number of Wins per Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create separate DataFrames for home wins and away wins
home_wins = data[data['home_team'] == data['winner']]
away_wins = data[data['away_team'] == data['winner']]

# Count the number of home wins and away wins for each team
home_wins_per_team = home_wins['winner'].value_counts()
away_wins_per_team = away_wins['winner'].value_counts()

# Plot the bar chart for each team showing the count of home and away wins
teams = home_wins_per_team.index.union(away_wins_per_team.index)
teams.sort_values()

plt.figure(figsize=(12, 6))
bar_width = 0.35
index = range(len(teams))

plt.bar(index, home_wins_per_team.reindex(teams, fill_value=0), bar_width, label='Home Wins', color='skyblue')
plt.bar(index, away_wins_per_team.reindex(teams, fill_value=0), bar_width, label='Away Wins', color='salmon', bottom=home_wins_per_team.reindex(teams, fill_value=0))

plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Home vs. Away Wins for Each Team')
plt.xticks(index, teams, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# # Filter the DataFrame to include only matches where the toss was won by each team
toss_wins = data[['toss_won', 'decision']].dropna()

# Count the occurrences of each toss decision (bat or bowl) for each team
toss_decisions_per_team = pd.crosstab(toss_wins['toss_won'], toss_wins['decision'])

# Plot the stacked bar chart for each team showing the distribution of toss decisions
toss_decisions_per_team.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Distribution of Toss Decisions (Bat or Bowl) by Teams')
plt.xticks(rotation=45)
plt.legend(title='Toss Decision')
plt.tight_layout()
plt.show()
```

```python
# Bar chart: Number of matches officiated by each umpire
plt.figure(figsize=(12, 6))
umpire_counts = pd.concat([data['umpire1'], data['umpire2']]).value_counts()
umpire_counts.plot(kind='bar', color='salmon')
plt.title('Number of Matches Officiated by Each Umpire')
plt.xlabel('Umpire')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart: Number of matches played at each venue
plt.figure(figsize=(12, 6))
data['venue_name'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Number of Matches Played at Each Venue')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Time series plot: Number of matches scheduled per match day
plt.figure(figsize=(12, 6))
data['match_days'].value_counts().plot(kind='line', marker='o', color='salmon')
plt.title('Number of Matches Scheduled per Match Day')
plt.xlabel('Match Day')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
